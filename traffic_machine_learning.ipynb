```python
import pandas as pd
```

```python
nc_traffic_data = pd.read_csv('incidents-3.csv')
nc_traffic_data
```

```python
nc_traffic_data = nc_traffic_data.drop(columns=[
    'Division', 'RoadNameSuffix', 'CommonName', 'Direction','CrossStreet', 'CrossStreetSuffix', 'CrossStreetCommonName',
    'EventName', 'InNearCity', 'MMFrom', 'MMto', 'StartLatitude', 'StartLongitude', 'EndLatitude', 
    'EndLongitude', 'IncidentID','LastUpdateDate (EST)', 'Detour'
])
```

```python
nc_traffic_data_I40 = nc_traffic_data.loc[(nc_traffic_data['RoadName'] == 'I-77')] 
```

```python
nc_traffic_data_I77 = nc_traffic_data.loc[(nc_traffic_data['RoadName'] == 'I-40')]
```

```python
nc_traffic_data_filtered = pd.concat([nc_traffic_data_I40,nc_traffic_data_I77])
```

```python
nc_traffic_data_filtered.set_index('County')
```

```python
interstate_segment_data= [
    ['Mecklenburg', 18.6], 
    ['Yadkin', 8.2],
    ['Iredell', 33.8], 
    ['Duplin', 20.9 ],
    ['Pender', 17.9],
    ['Sampson', 14.4],
    ['Wake', 12.5],
    ['Orange', 7.5],
    ['Davie', 12],
    ['Forsyth', 8.2],
    ['Buncombe', 28.3],
    ['Haywood', 33.2]
]
    
interstate_segment_data_df = pd.DataFrame(interstate_segment_data, columns =['County', 'Length(mi)'])
interstate_segment_data_df.set_index('County')
```

```python
nc_traffic_data_filtered_df = pd.merge(nc_traffic_data_filtered, interstate_segment_data_df).set_index('County')
nc_traffic_data_filtered_df
```

```python
nc_traffic_data_filtered_df.dtypes
```

```python
nc_traffic_data_filtered_df['StartTime (EST)'] = pd.to_datetime(nc_traffic_data_filtered_df['StartTime (EST)'])
nc_traffic_data_filtered_df['EndTime (EST)'] = pd.to_datetime(nc_traffic_data_filtered_df['EndTime (EST)'])
```

```python
nc_traffic_data_filtered_df.dtypes
```

```python
nc_traffic_data_filtered_df
```

```python

AvgCommuteTime = []

for value in nc_traffic_data_filtered_df["Length(mi)"]:
    value/70
    if value >= 0:
        AvgCommuteTime.append(value/70)
    elif value < 0 and value > 1:
        AvgCommuteTime.append("Invalid")
        
nc_traffic_data_filtered_df["AvgCommuteTime"] = AvgCommuteTime 
display(nc_traffic_data_filtered_df)
```
